```python
df=pd.read_excel("combined_mf.xlsx")
df.head()
```

```python
df=df.sort_values(by='Date')
```

```python
df.head()
```

Creating a new column named returns to examine which funds return the most profit 

Firstly, created two new dataframes: start_nav which records the start NAV date of all the funds, and end_nav which does the same for the end dates

```python
start_nav=df.groupby('Scheme Name').first().reset_index()
end_nav=df.groupby('Scheme Name').last().reset_index()
```

```python
start_nav=start_nav.rename(columns={
	"Net Asset Value" : "Start NAV",
	"Date": "Start date"
})
```

```python
start_nav
```

```python
end_nav=end_nav.rename(columns={
	"Net Asset Value":"End NAV",
	"Date":"End date"
})
```

```python
end_nav
```

Next, we merge the two dataframes into one so as to move on to calculating the returns

```python
nav_summary=start_nav.merge(end_nav,how="outer")
```

```python
nav_summary
```

Next, we add the "Returns%" column:

```python
nav_summary['Returns']=((nav_summary['End NAV']-nav_summary['Start NAV'])/nav_summary['Start NAV'])*100
nav_summary['Returns']=nav_summary['Returns'].round(2)
```

```python
nav_summary
```

```python
nav_summary.to_excel("nav_summary.xlsx",index=False)
```
